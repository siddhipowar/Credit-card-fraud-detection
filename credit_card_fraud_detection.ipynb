Importing the dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas frame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# display first 5 rows of the dataset
credit_card_data.head()
```

```python
#display last 5 rows of the dataset
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
credit_card_data['Class'].value_counts()
```

Dataset is unbalanced so undersampling is required.

Let 0 = normal transaction
Let 1 = fraudulent transactions

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

Under-sampling

Build a sample dataset containing similar distribution of normal transactions and fraudulent transactions

Number of Fradulent Transactions = 492

```python
legit_sample = legit.sample(n = 492)
```

Concatenate two data frames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and targets

```python
X = new_dataset.drop(columns = 'Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into train & test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model training

Logistic Regression


```python
model = LogisticRegression()
```

```python
#training the Logistic Regression model with Trianing data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data: ', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data: ', test_data_accuracy)
```
